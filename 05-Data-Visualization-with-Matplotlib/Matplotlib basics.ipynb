numpy : linspace : https://docs.scipy.org/doc/numpy/reference/generated/numpy.linspace.html

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2;
```

Matplotlib : plot : https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html

```python
# Functional
plt.plot(x,y)
# plt.show() -> for only direct plot
```

```python
# Plot with labels
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

## Multiplot 

_One plot containing many sub plot_ <br/>
SubPlot: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

#### Object Oriented

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)

axes1.set_title('Larger Plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[0].set_title("First Plot")
axes[1].plot(y,x)
axes[1].set_title("Second Plot")
# axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3)
plt.tight_layout()
```

## Figure Size and Dip

```python
# fig = plt.figure(figsize=(3,2),dpi=100)
fig = plt.figure(figsize=(8,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

Lengend: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3, label= 'X Cubed')
# ax.legend()
ax.legend(loc=0)
```

## Plot Appearance

Plot Color: https://matplotlib.org/2.1.1/api/_as_gen/matplotlib.pyplot.plot.html <br>
Line Style: https://matplotlib.org/2.1.2/api/_as_gen/matplotlib.pyplot.plot.html

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# ax.plot(x,y, color='orange')
# ax.plot(x,y, color='#2c3e50') # RGB Hex Code
# ax.plot(x,y, color='#2c3e50', linewidth=0.5)
# ax.plot(x,y, color='#2c3e50', linewidth=3, alpha=0.5)
# ax.plot(x,y, color='#2c3e50', lw=3, linestyle=":")
# ax.plot(x,y, color='#2c3e50', lw=3, linestyle="steps")
ax.plot(x,y, color='#2c3e50', lw=3, ls="--")
```

## Marker
Marker Link : https://matplotlib.org/api/markers_api.html

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# ax.plot(x,y, color='#2c3e50', lw=3, ls='-',marker='o')
ax.plot(x,y, color='#2c3e50', lw=1, ls='-',marker='o', markersize=20,
       markerfacecolor='yellow',markeredgewidth= 3, markeredgecolor='green')
```

## Plot Range

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# ax.plot(x,y, color='#2c3e50', lw=3, ls='-',marker='o')
ax.plot(x,y, color='#2c3e50', lw=2, ls='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
